```python
%%R
####Caso modelo del porcentaje, con epsilon = 0.8

#Vector de D's (probabilidades de reclamacion por individuo)
D<-c(rep(0.01,100), rep(0.01,60),rep(0.02,150), rep(0.02,40)) #cada probabilidad es independiente, es una variable Bernoulli cada una

#Simulas las reclamaciones con distribucion Bernoulli(Dj) (Binomial (1,Dj))

m<-100 #Establecer el numero de simular S
S<-0      #Inicializar S

for (i in 1:m){
  aux_D<-0
  aux_C<-0
  
  for (j in 1:length(D)){
    aux_D[j]<-rbinom(1,1,D[j])
  }
  #Si son distintas distribuciones de C, hay que hacer una simulacion para cada una
  aux_C<-c(runif(100,0,10),runif(60,10,20),runif(150,0,10), runif(40,10,20))
  
  S[i]<-sum(aux_D*aux_C)
}

hist(S, probability = TRUE) #Histograma de probabilidad
quantile(S,c(0, 0.2,0.8))
```
