```python
%%R
####################  RESTAURANTE #################### 
n= median(sample(15:20)) #posibles asistentes a la reunión
q=.5 #probabilidad de un asistente no pague su consumo. 
M=50000 #Tamaño de muestra de S
S=0 #variable auxiliar que guardara el posible valor de la posible cantidad total a pagar 
for(i in 1:M){
  D=rbinom(n,1,q) #caso particular 
  C=sample(700:800) # un valor aleatorio entre 700 y 800
  S[i]=sum(D*median(C))} 
hist(S,freq=FALSE,breaks=100, main="Distirbucion de S")
plot(density(S))
P = median(S) #Tomamos el promedio del riesgo como prima.
print(P)
```
