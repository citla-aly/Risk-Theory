**1**.- Sea Gamma(4, 7) y X2 una v. a. con distribución Gamma(16, 7). Grafique las 2−mezcla de gammas. Escriba un algoritmo claro y sencillo para muestrear n va's de esta distribución.

```python
%%R

############# Simulación de observaciones de la distribución 2-Gamma  ########


#Tamaño de la muestra 
m = 10000
a1 = 0.2
a2 = 0.8 

G <- rep(0,m)
g1 <- rgamma(m,4,7)
g2 <- rgamma(m,16,7)
for(i in 1:m){G[i] <-  a1*g1[i] + a2*g2[i]}

### Histograma para ver como se distribuyen los valores obtenidos después de n simulaciones  

hist(g1,freq=FALSE,breaks=100, main="Distribucion Gamm(4,7)  n=20")
lines(density(g1))

hist(g2,freq=FALSE,breaks=100, main="Distribucion Gamm(16,7)  n=20")
lines(density(g2))

hist(G,freq=FALSE,breaks=100, main="Distribucion 2-Gamma  n=20")
lines(density(G))
```

```python
%%R

####Caso 3.2: se tiene la distribuciÃ³n de las reclamaciones tipo DePril2, i.e. la distribuciÃ³n de reclamaciÃ³n ya considera el 0
#Con varios grupos de distribuciÃ³n

#Se deben tener los montos probables a reclamar y su probabilidad de reclamaciÃ³n, por grupo
montos_reclamo_1<-c(2000,0)
prob_reclamo_1<-c(.01, 1-0.01) #Aqui sÃ­ deben de sumar 1

n_1<-200 #Establecer el numero de sujetos

#Se deben tener los montos probables a reclamar y su probabilidad de reclamaciÃ³n, por grupo
montos_reclamo_2<-c(2000,0)
prob_reclamo_2<- c(0.1, 1-0.1) #Aqui sÃ­ deben de sumar 1

n_2<-500 #Establecer el numero de sujetos

#Se deben tener los montos probables a reclamar y su probabilidad de reclamaciÃ³n, por grupo
montos_reclamo_3<-c(2000,0)
prob_reclamo_3<-c(0.02, 1-0.02) #Aqui sÃ­ deben de sumar 1

n_3<-120 #Establecer el numero de sujetos

montos_reclamo_4<-c(2000,0)
prob_reclamo_4<-c(0.005, 1-0.005) #Aqui sÃ­ deben de sumar 1

n_4<-180 #Establecer el numero de sujetos
#Establecer el numero de simulaciones
m<-10000
S<-0 #Inicializar 0

for (j in 1:m){
  #Construir tantos auxiliares como grupos
  
  aux_1<-sum(sample(montos_reclamo_1,n_1,replace=TRUE,prob=prob_reclamo_1))
  aux_2<-sum(sample(montos_reclamo_2,n_2,replace=TRUE,prob=prob_reclamo_2))
  aux_3<-sum(sample(montos_reclamo_3,n_3,replace=TRUE,prob=prob_reclamo_3))
  aux_4<-sum(sample(montos_reclamo_3,n_4,replace=TRUE,prob=prob_reclamo_4))
  S[j]<-aux_1+aux_2+aux_3 + aux_4
}

hist(S, probability = TRUE) #Histograma de probabilidad
#200, 910, 000
mean(S)
var(S)
```
